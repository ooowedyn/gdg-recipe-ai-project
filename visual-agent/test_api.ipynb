```python
import os
import requests
from dotenv import load_dotenv

load_dotenv()

FOOD_API_KEY = os.getenv("FOOD_API_KEY")

url = f"http://openapi.foodsafetykorea.go.kr/api/{FOOD_API_KEY}/COOKRCP01/json/1/500"

response = requests.get(url)
data = response.json()

rows = data["COOKRCP01"]["row"]

print("가져온 개수:", len(rows))
print("번호\t일련번호\t메뉴명")

for idx, row in enumerate(rows, start=1):
    print(f"{idx}\t{row.get('RCP_SEQ', '')}\t{row.get('RCP_NM', '')}")
```
